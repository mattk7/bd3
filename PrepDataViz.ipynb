```python
%pylab inline
import numpy as np
import pandas as pd
import json
```

```python
fname = "Data/WBData_Clean.csv"
with open(fname,'r') as fp:
    df = pd.read_csv(fp)
fname = "Data/olympics_2012_Clean.csv"
with open(fname,'r') as fp:
    df_2 = pd.read_csv(fp)
```

```python
Country_names  = df['Country Name'].unique()
Var_names = df['﻿Series Name'].unique()
Vars = ['Internet users (per 100 people)',\
        'GDP (current US$)',\
        'GDP per capita (current US$)',\
        'GNI per capita, Atlas method (current US$)',\
        'GNI per capita, PPP (current international $)',\
        'GINI index',\
        'Population (Total)'
]
```

```python
year = 2012
data_dict = dict( (name,dict()) for name in Country_names )
for var in Vars:
    for name in Country_names:
        value = df[(df['Country Name'] == name) & (df['﻿Series Name'] == var)]['2001'].values
        data_dict[name][var] = value[0] if value else np.nan
```

```python
data_dict
```

```python
df_2 = df_2.replace(to_replace = np.nan, value = 0.0)
df_2['Sex'] = df_2['Sex'].apply(lambda x:x.lower())

df_2['Athletes'] = 1.0
df_2['Female_Athletes'] = df_2['Sex']=='f'
df_2['Male_Athletes'] = df_2['Sex']=='m'

df_2['Female_Gold'] = df_2['Female_Athletes']*df_2['Gold']
df_2['Male_Gold'] = df_2['Male_Athletes']*df_2['Gold']

df_2['Female_Silver'] = df_2['Female_Athletes']*df_2['Silver']
df_2['Male_Silver'] = df_2['Male_Athletes']*df_2['Silver']

df_2['Female_Bronze'] = df_2['Female_Athletes']*df_2['Bronze']
df_2['Male_Bronze'] = df_2['Male_Athletes']*df_2['Bronze']

df_2['Medals'] = df_2['Gold'] + df_2['Silver'] + df_2['Bronze']
df_2['Female_Medals'] = df_2['Female_Athletes']*(df_2['Gold'] + df_2['Silver'] + df_2['Bronze'])
df_2['Male_Medals'] = df_2['Male_Athletes']*(df_2['Gold'] + df_2['Silver'] + df_2['Bronze'])

df_2['Medalists'] = df_2['Medals'] > 0
df_2['Female_Medalists'] = df_2['Female_Medals'] > 0
df_2['Male_Medalists'] = df_2['Male_Medals'] > 0

df_2.head()
```

```python
grouped_data = df_2[['Country','Medals','Gold','Silver','Bronze','Medalists','Athletes',\
                     'Female_Medals','Female_Gold','Female_Silver','Female_Bronze','Female_Medalists','Female_Athletes',\
                     'Male_Medals','Male_Gold','Male_Silver','Male_Bronze','Male_Medalists','Male_Athletes']].groupby('Country').agg(np.sum)
```

```python
grouped_data.head()
```

```python
for country in data_dict:
    data_dict[country].update(grouped_data.stack()[country])
```

```python
with open('./Data/data_joined.json', 'w') as fp:
    json.dump(data_dict,fp)
```

```python
df = pd.DataFrame.from_dict(data_dict, orient = 'index')
```

```python
df['Medals_TeamSize'] = df['Medals']/df['Athletes']
df['Female_Medals_TeamSize'] = df['Female_Medals']/df['Female_Athletes']
df['Male_Medals_TeamSize'] = df['Male_Medals']/df['Male_Athletes']

df['Medalists_TeamSize'] = df['Medalists']/df['Athletes']
df['Female_Medalists_TeamSize'] = df['Female_Medalists']/df['Female_Athletes']
df['Male_Medalists_TeamSize'] = df['Male_Medalists']/df['Male_Athletes']

df['Medals_Medalists'] = df['Medals']/df['Medalists']
df['Female_Medals_Medalists'] = df['Female_Medals']/df['Female_Medalists']
df['Male_Medals_Medalists'] = df['Male_Medals']/df['Male_Medalists']

df['Medals_Pop'] = df['Medals']/df['Population (Total)']
df['Female_Medals_Pop'] = df['Female_Medals']/df['Population (Total)']
df['Male_Medals_Pop'] = df['Male_Medals']/df['Population (Total)']

df['Medals_logPop'] = df['Medals']/np.log(df['Population (Total)'])
df['Female_Medals_logPop'] = df['Female_Medals']/np.log(df['Population (Total)'])
df['Male_Medals_logPop'] = df['Male_Medals']/np.log(df['Population (Total)'])

df['Medalists_Pop'] = df['Medalists']/df['Population (Total)']
df['Female_Medalists_Pop'] = df['Female_Medalists']/df['Population (Total)']
df['Male_Medalists_Pop'] = df['Male_Medalists']/df['Population (Total)']

df['Medalists_logPop'] = df['Medalists']/np.log(df['Population (Total)'])
df['Female_Medalists_logPop'] = df['Female_Medalists']/np.log(df['Population (Total)'])
df['Male_Medalists_logPop'] = df['Male_Medalists']/np.log(df['Population (Total)'])

df['Athletes_Pop'] = df['Athletes']/df['Population (Total)']
df['Female_Athletes_Pop'] = df['Female_Athletes']/df['Population (Total)']
df['Male_Athletes_Pop'] = df['Male_Athletes']/df['Population (Total)']

df['Athletes_logPop'] = df['Athletes']/np.log(df['Population (Total)'])
df['Female_Athletes_logPop'] = df['Female_Athletes']/np.log(df['Population (Total)'])
df['Male_Athletes_logPop'] = df['Male_Athletes']/np.log(df['Population (Total)'])


```
